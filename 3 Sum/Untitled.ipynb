```python
class Solution:
    def threeSum(self, nums):
        length=len(nums)
        nums=sorted(nums)
        res=[]
        for idx,val in enumerate(nums):
            if idx>=length-2:
                break
            if idx>0 and val==nums[idx-1]:
                continue
            if val>0:
                break
            l=idx+1
            r=length-1
            while l<r:
                summ=val+nums[l]+nums[r]
                print(summ,l,r)
                if summ==0:
                    res.append((val,nums[l],nums[r]))
                    l+=1
                    r-=1
                if summ<0:
                    l+=1
                else:
                    l-=1
        return res
                    
                
```
